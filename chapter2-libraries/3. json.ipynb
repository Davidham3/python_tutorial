python提供了一个叫json的库专门用于处理json字符串。

```python
import json
```

假设我们有一个列表，里面有一个字典，我们想把它转换成json字符串，使用json.dumps

```python
test_data = [{ 'a': 1, 'b': 2, 'c': 3, 'd': 4, 'e': 5 }]
```

```python
json_str = json.dumps(test_data)
```

```python
print(json_str)
print(type(json_str))
```

现在想解析json字符串，使用json.loads

```python
new_data = json.loads(json_str)
```

```python
print(new_data)
```

一般常用的就是这两个函数，其他的可见help(json)

```python
help(json)
```
